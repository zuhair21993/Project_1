```python
nj_asthma_hosp = nj_asthma_df.iloc[8:17, [1,-1]]
nj_asthma_hosp_renamed = nj_asthma_hosp.rename(columns={"Asthma Hospitalizations Per Year": "NJ Asthma Hospitalizations"})
nj_asthma_hosp_final = nj_asthma_hosp_renamed.reset_index(drop=True)
nj_asthma_hosp_final
```

```python
ct_copd_file = pd.read_csv("../Output/CT COPD Hosp 2006 -2015.csv")
cleaned_df = ct_copd_file.dropna(how="any")
short_df = cleaned_df.iloc[1:, [0,2]]
ct_copd_hosp_final = short_df.rename(columns={"Unnamed: 0": "Year", "Both Sexes": "CT COPD Hospitalizations"})
ct_copd_hosp_final.reset_index(drop=True)
```

```python
#concatenating Asthma Hospitalizations
ct_asthma_hosp_final = ct_asthma_hosp_final.apply(pd.to_numeric)
ny_asthma_hosp_final = ny_asthma_hosp_final.apply(pd.to_numeric)
merged1 = pd.merge(nj_asthma_hosp_final, ct_asthma_hosp_final, how='outer').sort_values("Year")
compiled_asthma_hosp = pd.merge(merged1, ny_asthma_hosp_final)
compiled_asthma_hosp
```

```python
#concatenating Asthma ED Visits
ct_asthma_ed_final = ct_asthma_ed_final.apply(pd.to_numeric)
ny_asthma_ed_final = ny_asthma_ed_final.apply(pd.to_numeric)
merged2 = pd.merge(nj_asthma_ed_final, ct_asthma_ed_final, how='outer').sort_values("Year")
compiled_asthma_ed = pd.merge(merged2, ny_asthma_ed_final)
compiled_asthma_ed
```

```python
#concatenating COPD Hospitalizations
ct_copd_hosp_final = ct_copd_hosp_final.apply(pd.to_numeric)
ny_copd_hosp_final = ny_copd_hosp_final.apply(pd.to_numeric)
nj_copd_hosp_final = nj_copd_hosp_final.apply(pd.to_numeric)
merged3 = pd.merge(nj_copd_hosp_final, ct_copd_hosp_final, how='outer').sort_values("Year")
compiled_copd_hosp = pd.merge(merged3, ny_copd_hosp_final)
compiled_copd_hosp
```

```python
#Plot for Asthma Hospitalizations 
years = np.arange(2006, 2017, 1)
nj1 = compiled_asthma_hosp["NJ Asthma Hospitalizations"]
ct1 = compiled_asthma_hosp["CT Asthma Hospitalizations"]
ny1 = compiled_asthma_hosp["NY Asthma Hospitalizations"]
yr = [year for year in years]
plt.figure(figsize=(8,6))
plt.bar(yr, nj1, bottom=ct1+ny1, label="New Jersey", alpha=0.75, edgecolor="black", color="blue")
plt.bar(yr, ct1, bottom=ny1, label="Connecticut", alpha=0.75, edgecolor="black", color="red")
plt.bar(yr, ny1, label="New York", alpha=0.75, edgecolor="black", color="green")

plt.ylabel("No of People Hospitalized")
plt.xlabel("Years")
plt.title("Tri-State Asthma Hospitalizations")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
#Plot for Asthma ED visits 
years = np.arange(2006, 2017, 1)
nj2 = compiled_asthma_ed["NJ Asthma ED Visits"]
ct2 = compiled_asthma_ed["CT Asthma ED Visits"]
ny2 = compiled_asthma_ed["NY Asthma ED Visits"]
yr = [year for year in years]
plt.figure(figsize=(8,6))
plt.bar(yr, nj2, bottom=ct2+ny2, label="New Jersey", alpha=0.75, edgecolor="black", color="blue")
plt.bar(yr, ct2, bottom=ny2, label="Connecticut", alpha=0.75, edgecolor="black", color="red")
plt.bar(yr, ny2, label="New York", alpha=0.75, edgecolor="black", color="green")

plt.ylabel("No of ED Visits")
plt.xlabel("Years")
plt.title("Tri-State Asthma ED Visits")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
