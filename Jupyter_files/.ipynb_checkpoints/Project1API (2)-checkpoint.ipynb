```python
# Dependencies
import requests
import json
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
```

```python
# URL for GET requests to retrieve vehicle data
CT_url = "https://data.cdc.gov/resource/cjae-szjv.json?statename=Connecticut&measureid=84"
```

```python
# Retrieving data and converting it into JSON
print(requests.get(CT_url).json())
```

```python
# Pretty Print the output of the JSON
CT_response = requests.get(CT_url).json()
print(json.dumps(CT_response, indent=4, sort_keys=True))
```

```python
Connecticut_df = pd.read_json("https://data.cdc.gov/resource/cjae-szjv.json?statename=Connecticut&measureid=84")
Connecticut_df
```

```python
Connecticut_2013 = Connecticut_df.loc[Connecticut_df["reportyear"] == 2013]
Connecticut_2013
```

```python
CT = Connecticut_2013.groupby("countyname",sort=True).sum()
CT_allcolumns = CT.reset_index()
CT_allcolumns
```

```python
CT_2013 = CT_allcolumns.drop(["measureid", "statefips", "countyfips", "monitoronly", "reportyear"], axis=1)
CT_2013
```

```python
cols = 1, 2, 1, 2, 1, 2, 1, 2
CT_2013['columns'] = cols
rows = 1, 1, 2, 2, 3, 3, 4, 4
CT_2013['rows'] = rows
CT_2013
```

```python
CT_County_Names = ((np.asarray(CT_2013['countyname'])).reshape(4,2))
CT_Person_Days = ((np.asarray(CT_2013['value'])).reshape(4,2))

print(CT_County_Names)
print(CT_Person_Days)
```

```python
result_CT = CT_2013.pivot(index='rows', columns='columns', values='value')
result_CT
```

```python
CTlabels = (np.asarray(["{0} \n {1:.2f}".format(conn_counties, ctperson_days)
                     for conn_counties, ctperson_days in zip(CT_County_Names.flatten(),
                                           CT_Person_Days.flatten())])
         ).reshape(4,2)
CTlabels
```

```python
fig, ax = plt.subplots(figsize=(20,7))
plt.title("Heat Map of Connecticut Person-Days with 8 hour or More Maximum O2 Concentration Exceeding NAAQS Standard (2013 By County)")
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
sb.heatmap(result_CT,annot=CTlabels,fmt="",cmap='RdYlBu_r',linewidths=0,ax=ax)
plt.show()
```
