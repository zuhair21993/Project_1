```python
import os
import glob
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import requests
import zipfile
```

```python
# Concatenating AQI Files for the last 40 years
#https://www.youtube.com/watch?v=KoRT-v0SzMs

def concatenate(indir, outfile):
    os.chdir(indir)
    fileList = glob.glob("*.csv")
    dfList = []
    for filename in fileList:
        #print(filename) "commented out to avoid being printed on terminal"
        df = pd.read_csv(filename)
        dfList.append(df)
    concatDF = pd.concat(dfList, axis=0)
    concatDF.to_csv(outfile, index=None)
    
aqi_dir = "C:/Users/sudha/Documents/DataViz/Project _1/Still_Breathing-Project_1-/Resources/annual_aqi_county"
aqi_output = "C:/Users/sudha/Documents/DataViz/Project _1/Still_Breathing-Project_1-/Output/aqi_compiled.csv"
concatenate(aqi_dir, aqi_output)
```

```python
aqi_complete_df = pd.read_csv(aqi_output)
aqi_complete_df.head()
```

```python
# create new columns for acceptable air quality days and poor air quality days
aqi_complete_df["Acceptable AQ Days"] = aqi_complete_df["Good Days"] + aqi_complete_df["Moderate Days"]

aqi_complete_df["Poor AQ Days"] = aqi_complete_df["Unhealthy for Sensitive Groups Days"] + aqi_complete_df["Unhealthy Days"]\
+ aqi_complete_df["Very Unhealthy Days"] + aqi_complete_df["Hazardous Days"]
aqi_complete_df.head()
```

```python
# Pull out NY, NJ, CT only
aqi_tristate_df = aqi_complete_df.loc[aqi_complete_df["State"].isin(["Connecticut", "New Jersey", "New York"])]
aqi_tristate_df.head()
```

```python
# Create 2 new columns for percent acceptable aq days and percent poor aq days
aqi_tristate_df["Percent Acceptable AQ Days"] = (aqi_tristate_df["Acceptable AQ Days"]/aqi_tristate_df["Days with AQI"])*100
aqi_tristate_df["Percent Poor AQ Days"] = (aqi_tristate_df["Poor AQ Days"]/aqi_tristate_df["Days with AQI"])*100
aqi_tristate_df.head()
```

```python
# Round % to 2 decimal places
aqi_tristate_df.round({"Percent Acceptable AQ Days":2, "Percent Poor AQ Days":2})
```

```python
# Plot trend in AQ over 40 years for tri-state area

fig, ax = plt.subplots(figsize=(10,8))
x_axis = aqi_tristate_df["Year"]
plt.ylim(-5,105)
plt.scatter(x_axis, aqi_tristate_df["Percent Acceptable AQ Days"], color='green', alpha=0.75, edgecolor='black', \
            s=50, label='% Acceptable AQ Days')
plt.scatter(x_axis, aqi_tristate_df["Percent Poor AQ Days"], color='red', alpha=0.75, edgecolor='brown', s=50, \
            label = '% Poor AQ Days')
plt.xlabel("Year", fontstyle='italic')
plt.ylabel("Acceptable or Poor Air Quality Days(%)", fontstyle='italic')
plt.title("Trends in Acceptable & Poor Air Quality Days by AQI Index \n (Tristate Area 1980-2019)", fontweight='bold')
plt.legend()
plt.show()
```
